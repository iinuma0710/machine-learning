# 第1章 大規模言語モデルを理解する

Open AI の ChatGPT をはじめ、多くの大規模言語モデル (Learge Language Model: LLM) が登場し、自然言語処理 (Natural Language Processing: NLP) は新たな時代を迎えた。
LLM の成功は、近年登場した Transformer アーキテクチャと、学習に用いられる膨大なデータによってもたらされたものである。
本書では、Transformer アーキテクチャに基づく ChatGPT 型の LLM を段階的に実装しながら、LLM を理解することを目標とする。
