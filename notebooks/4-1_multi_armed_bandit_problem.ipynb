# 4.1 多腕バンディット問題 (Multi-armed bandit problem)

## 4.1.1 k 本腕バンディット問題
$k$ 個の異なる選択肢の中からいずれか1つの行動を選択した時に、その行動に依存する定常確率分布から数値的な報酬が発生するケースを考える。
この選択を繰り返し行った時に、ある時間間隔または時間ステップにおいて、得られる報酬の合計の期待値を最大化する問題を、**$k$ 本腕バンディット問題**と呼ぶ。
バンディットとはスロットマシンの別名であり、選択は $k$ 台のスロットマシンの中からレバーを引くマシンを選ぶこと、行動は選択したスロットマシンのレバーを引くことにあたる。
報酬は選択したスロットマシンで得られる賭けの利益である。
繰り返しスロットマシンを回して最良のマシンを見つけ出し、賭けの利益を最大化することが目標となる。
以降では、$k$ 個の行動のそれぞれに対して設定された報酬の期待値、または平均値を、その行動の**価値 (value)** と呼ぶこととする。
ある時間ステップで選択された行動を $A_t$、それに対応する報酬を $R_t$ とする。
ある任意の行動 $a$ の価値 $q_*(a)$ は、$a$ が選択された時に得られる期待報酬であり、次のように定義される。

$$
q_*(a) \triangleq \mathbb{E} \big[ R_t | A_t = a \big]
$$

$q_*(a)$ は分からないが推定することは可能である時、ある時間ステップ $t$ において推定された行動 $a$ の価値を $Q_t(a)$ とする。
この $Q_t(a)$ を $q_*(a)$　に近づけることを考える。  
ある時間ステップにおいて、行動価値の推定値が最大となる行動を**グリーディ (貪欲) 行動**と呼ぶ。
この行動を選択した場合、行動価値について現在持っている知識を**活用 (exploit)** しているという。
一方、グリーディでない行動を選択した場合には、そのほかの行動の価値推定を改善できるため、**探索 (exploring)** しているという。
グリーディな行動を選択することで1つの行動によって得られる報酬の期待値を最大化することができる一方、長期的には探索によってより高い合計報酬が得られる可能性がある。
そのため、先に多くの時間ステップが残されている状況であれば、短期的には得られる報酬が減少したとしてもグリーディな行動以外を探索して、より良い行動を見つけ出すのが得策かもしれない。
1つの行動選択の中で探索と活用を両立することはできないため、この関係を「探索と活用のトレードオフ」と呼ぶ。  
これまで、定常性や事前知識に強い仮定を置いた上で、探索と活用のバランスの取り方について様々な手法が提案されている。
しかし、その仮定が崩れてしまっていたり、そもそも崩れているかの確認が困難であったりする場合には、それらの手法が適用できなくなってしまう。
ここでは、複雑な仮定は置かず、簡単に探索と活用のバランスをとることだけを考え、活用を続けるよりも探索を行うことでより良い結果が得られることを確認する。
